# Superdense Coding: Interactive Exploration
#### Superdense coding is a quantum communication protocol that allows the transmission of two classical bits of information using only one qubit. This notebook provides an interactive demonstration of superdense coding, allowing you to experiment with different encodings and observe the results.


#### **Setup**

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from IPython.display import display

def create_circuit(initial_pair):
    q = QuantumRegister(2, 'qreg')
    c = ClassicalRegister(2, 'creg')
    qc = QuantumCircuit(q, c)
    return qc, q, c

def run_circuit(qc):
    job = AerSimulator().run(qc)
    result = job.result()
    counts = result.get_counts(qc)
    return counts

def display_results(qc, initial_pair, counts):
    print('Classical Pair:', initial_pair)
    print('Measurement:', counts)
    display(qc.draw(output='mpl', reverse_bits=True))
```

## Part 1: Basic Superdense Coding
#### Let's start with a basic example of superdense coding.
##### In this example, we're encoding the classical pair "10". The Z gate applied to the second qubit corresponds to this encoding. Run the cell and observe the results. Does the measurement match the initial pair?

```python
initial_pair = "10"
qc, q, c = create_circuit(initial_pair)

# Encoding
qc.h(1)  # Apply Hadamard gate to the second qubit
qc.cx(1, 0)  # Apply CNOT gate to entangle the qubits
qc.barrier()  # Add a barrier for visual separation

# Apply additional operations based on the initial pair
qc.z(1)
qc.barrier()

# Decoding
qc.cx(1, 0)  # Apply CNOT gate again
qc.h(1)  # Apply Hadamard gate to the second qubit

# Measure the qubits
qc.measure(q, c)

counts = run_circuit(qc)
display_results(qc, initial_pair, counts)
```
